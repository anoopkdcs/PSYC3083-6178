<b><i><u>Example #1:</b></i></u> 
Develop an algorithm to search a number in a list of numbers.  
From the list [2, 5, 8, 12, 16, 23, 38, 56, 72, 91] search and identify the number 23 


<b><i><u> Solution #1: Linear Search </b></i></u> </br>
Step 1: Start from the 0th element of the input, compare the key value with the value present in the 0th  index </br>
Step 2: If the value matches with the key, we found the value  </br>
Step 3: If the value does not match with the key, compare the next element in the input  </br>
Step 4: Repeat Step 3 until there is a match found </br>
Step 5: If it is an unsuccessful search show message ‘number not found’ </br>

```python
def linear_search(input_data, key):
    count = 0
    n = len(input_data)
    for i in range(n):
        if(input_data[i] == key):
            print("\n -----> Program Output: The element is found at position", (i+1))
            print("\n")
            count = count + 1
    
    if(count == 0):
        print("\n \n -----> Program Output: The element is not present in the array")
        print("\n")


        
        
input_data = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
key = 23
linear_search(input_data, key)
```

<b><i><u> Solution #2: Binary Search  </b></i></u> </br>
Step 1: Select the middle item in the input and compare it with the key value to be searched. If it is matched “we   found the number” </br>
Step 2: If it does not match the key value, check if the key value is either greater than or less than the median value </br>
Step 3: </br>
&nbsp; &nbsp; &nbsp; &nbsp; If the key is greater, perform the search in the right sub-list; </br>
&nbsp; &nbsp; &nbsp; &nbsp; if the key is lower than the median value, perform the search in the left </br>
Step 4: Repeat Steps 1, 2 and 3 iteratively, until the size of sub-list becomes 1</br>
Step 5: If it is an unsuccessful search show message ‘number not found’</br>


```python
def binary_search(input_data, low, high, key):
    mid = (low + high) // 2
    
    if (low <= high):
        if(input_data[mid] == key):
            print("\n -----> Program Output: The element is found at position", mid+1)
            print("\n")

        elif(key < input_data[mid]):
            binary_search(input_data, low, mid-1, key)
        elif (input_data[mid] < key):
            binary_search(input_data, mid+1, high, key)
            
    if(low > high):
        print("\n \n -----> Program Output: The element is not present in the array")
        print("\n")


        
        
input_data = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
key = 91

low = 0
high = len(input_data)-1

binary_search(input_data, low, high, key)
```
